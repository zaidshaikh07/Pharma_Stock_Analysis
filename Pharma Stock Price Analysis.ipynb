```python
import pandas_datareader.data as web
import datetime as dt
import numpy
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
start = dt.datetime(2018, 1, 1)
end = dt.datetime.now()
```

```python
reddy = web.DataReader("DRREDDY.NS",'yahoo',start, end)
sun = web.DataReader("SUNPHARMA.NS",'yahoo',start, end)
gland = web.DataReader("GLAND.NS",'yahoo',start, end)
cipla = web.DataReader("CIPLA.NS",'yahoo',start, end)
```

```python
reddy.to_csv("Reddy_Stock.csv")
sun.to_csv("Sun_Stock.csv")
gland.to_csv("Gland_Stock.csv")
cipla.to_csv("Cipla_Stock.csv")
```

# Daily Percentage 
$$  r_t = {{p_t}\over{p_t}-1 } -1   $$

```python
reddy['returns'] = (reddy['Close']/reddy['Close'].shift(1))-1
reddy.head()
```

```python
sun['returns'] = (sun['Close']/sun['Close'].shift(1))-1
sun.head()
```

```python
gland['returns'] = (gland['Close']/gland['Close'].shift(1))-1
gland.head()
```

```python
cipla['returns'] = (cipla['Close']/cipla['Close'].shift(1))-1
cipla.head()
```

```python
reddy['returns'].hist(bins=50,label = 'Reddy',alpha = 0.3,figsize = (13,6))
sun['returns'].hist(bins=50,label = 'Sun',alpha = 0.3)
gland['returns'].hist(bins=50,label = 'gland',alpha = 0.3)
cipla['returns'].hist(bins=50,label = 'cipla',alpha = 0.3)
plt.legend()
```

```python
reddy['returns'].plot(kind='kde',label = 'Reddy',figsize = (17,6))
sun['returns'].plot(kind='kde',label = 'Sun')
gland['returns'].plot(kind='kde',label = 'Gland')
cipla['returns'].plot(kind='kde',label = 'Cipla')
plt.legend()
                      
```

```python
box_df = pd.concat([reddy['returns'],sun['returns'],cipla['returns']],axis = 1)
box_df.columns = ['Reddy Returns','Sun Returns','Cipla Returns']
box_df.plot(kind='box',figsize = (16,7))
```

```python
scatter_matrix(box_df,figsize = (8,8),hist_kwds={'bins':50},alpha = 0.6)
```

```python
reddy['Cumulative_Return'] = (1 + reddy['returns']).cumprod()
reddy.tail()
```

```python
sun['Cumulative_Return'] = (1 + sun['returns']).cumprod()
sun.tail()
```

```python
gland['Cumulative_Return'] = (1 + gland['returns']).cumprod()
gland.tail()
```

```python
cipla['Cumulative_Return'] = (1 + cipla['returns']).cumprod()
cipla.tail()
```

```python
reddy['Cumulative_Return'].plot(label = 'Reddy',figsize = (15,7))
sun['Cumulative_Return'].plot(label = 'Sun')
gland['Cumulative_Return'].plot(label = 'Gland')
cipla['Cumulative_Return'].plot(label = 'Cipla')                               
plt.title('Cumulative Return vs Time')
plt.legend()
```
